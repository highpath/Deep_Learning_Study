# 2. Importing data

```
import pandas as pd
```

```
train = pd.read_csv('data/train.csv', index_col=0)
test = pd.read_csv('data/test_x.csv', index_col=0)
submission = pd.read_csv('data/sample_submission.csv', index_col=0)
```

```
train_x = train.drop('voted', axis=1)
train_y = train['voted']
```

# 3. One hot encoding
- age_group:

```
print(set(train['age_group']))
print(set(train['gender']))
print(set(train['race']))
print(set(train['religion']))
```

```
train_x = pd.get_dummies(train_x)
test = pd.get_dummies(test)
```

```
import lightgbm as lgbm
```

```
model = lgbm.LGBMClassifier(n_estimators=500)
model.fit(train_x, train_y)
```

```
pred_y = model.predict(test)
```

```
pred_y
```

```
submission['voted']=pred_y
```

```
submission.to_csv('data/Baseline_01.csv')
```
